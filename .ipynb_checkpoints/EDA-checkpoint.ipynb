```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style='ticks')
```

```python
prescriber_df = pd.read_csv('medicare-part-d-opioid-prescriber-summary-file-2016.csv')
```

```python
prescriber_df.info()
```

```python
(prescriber_df.groupby(prescriber_df['Specialty Description'])['Opioid Claim Count'].mean() * 100).sort_values().plot(kind='barh',
                                                                                                                     figsize=(10,45))
```

```python
prescriber_df.groupby(['Specialty Description'])['Opioid Prescribing Rate'].mean().sort_values().plot(kind='barh',
                                                                                                   figsize=(10,45))
```

```python
prescriber_df.groupby(['NPPES Provider State'])['Opioid Claim Count'].count().sort_values().plot(kind='barh',
                                                                                                 figsize=(10,20))
```
